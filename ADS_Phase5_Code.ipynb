## Creating an ML Model

```python
#Dropping unwanted columns from the dataset
df = c_data_new.copy()
df = df.drop(['vaccines','date'],axis = 1)
```

```python
#Importing necessary packages to perform model training
from sklearn.preprocessing import LabelEncoder 
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
#Converting the categorical columns into numerical columns
le=LabelEncoder()
df['iso_code']=le.fit_transform(df['iso_code'])
```

```python
df['country']=le.fit_transform(df['country'])
```

```python
#Initializing the model
model = xgb.XGBRegressor(
    n_estimators=500,
    learning_rate=0.08,
    subsample=0.75,
    colsample_bytree=1,
    max_depth=7,
    gamma=0)
```

```python
X = df.drop('total_vaccinations', axis = 1)
y = df['total_vaccinations']
```

```python
#splitting the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
#fitting the data onto the model
model.fit(X_train, y_train)
```

```python
# getting the model predictions
y_pred = model.predict(X_test)
```

```python
#calculating the r2_score for the model prediction
score = np.sqrt(r2_score(y_test, y_pred))
```

```python
print(score) # the score is ~100 which means the model has 100% accuracy
```

```python
#plotting the data against the predictions
sns.scatterplot(x = y_test, y = y_pred)
```

```python
#getting the feature importances from the model
feat_imp = model.feature_importances_
```

```python
feature_imp = pd.Series(data = feat_imp, index = X.columns)
feature_imp
```
