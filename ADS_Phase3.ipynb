# Data Pre - Processing

```python
#removing the missing values
c_data_cleaned = c_data.dropna()

c_data_cleaned.info()

c_data_cleaned.describe()
```

```python
# Dropping source name and website from the data
c_data_new = c_data_cleaned.drop(['source_name','source_website'],axis = 1)
c_data_new.head()
```

```python
#Changing date column to date-type
c_data_new["date"] = pd.to_datetime(c_data_new["date"], format = '%Y-%m-%d')
```

```python
#Reset index
c_data_new.reset_index(drop=True, inplace=True)
```
